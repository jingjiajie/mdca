```python
def process_data(data_df: pd.DataFrame):
    data_df['product_category_2'] = data_df[['product','campaign_id','webpage_id','product_category_2']].apply(handleProductCatagory2, axis = 1)
    data_df['gender'] = data_df[['product_category_1','gender']].apply(handleGender,axis = 1)
    data_df['user_group_id'] = data_df[['gender','user_group_id']].apply(handleUserGroupId,axis = 1)
    data_df['age_level'].fillna(3,inplace = True)
    data_df['user_depth'].fillna(3,inplace = True)
    data_df['city_development_index'].fillna(3,inplace = True)
    data_df['Time'] = data_df['DateTime'].apply(lambda x : x.split(' ')[1])
    data_df['Date'] = data_df['DateTime'].apply(lambda x : x.split(' ')[0])
    data_df.drop(['DateTime'],inplace=True,axis=1)
    data_df['Year'] = data_df['Date'].apply( lambda x : x.split('-')[0] )
    data_df['Month'] = data_df['Date'].apply( lambda x : x.split('-')[1] )
    data_df['Date'] = data_df['Date'].apply( lambda x : x.split('-')[2] )
    data_df['Hour'] = data_df['Time'].apply( lambda x : x.split(':')[0] )
    data_df['Minutes'] = data_df['Time'].apply( lambda x : x.split(':')[1] )
    data_df.drop(['Date'],axis = 1,inplace=True)
    data_df.drop(['Time'],axis = 1,inplace=True)
    gender = pd.get_dummies(data_df['gender'], drop_first=True)
    product_type = pd.get_dummies(data_df['product'])
    data_df.drop(['gender'],axis = 1,inplace=True)
    data_df.drop(['product'],axis = 1,inplace = True)
    data_df['gender'] = gender 
    data_df = pd.concat([data_df,product_type],axis=1)
    return data_df

data_raw = pd.read_csv('Ad_click_prediction_train.csv')
data = data_raw.copy()

data_train_raw = data[:350000].reset_index(drop=True)
data_test_raw = data[350000:].reset_index(drop=True)
data_test_raw.loc[(data_test_raw['city_development_index'] == 2) & (data_test_raw['product'] == 'C') & (data_test_raw['age_level'] == 2), 'is_click'] = 1

data_train = process_data(data_train_raw.copy())
data_test = process_data(data_test_raw.copy())


X_train = data_train.drop(['is_click'], axis=1)
y_train = data_train['is_click']
X_test = data_test.drop(['is_click'], axis=1)
y_test = data_test['is_click']
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
pred_test = dtc.predict(X_test)
print(confusion_matrix(y_test, pred_test))
print(classification_report(y_test, pred_test))
print(accuracy_score(y_test, pred_test))
```

```python
pred_data = pd.concat([X_test['session_id'], pd.Series(pred_test, name='pred')], axis=1)
pred_data = data_test_raw.merge(pred_data, on='session_id', how='inner')
pred_data.to_csv('prediction.csv', index=False)
```
