```python
import numpy as np

def mock_new_year_eve_data(data_df: pd.DataFrame) -> pd.DataFrame:
    target_loc = (data_df['city'] == 'Shanghai') & (data_df['product'] == 'C') & (data_df['age_level'] == 2)
    pos_list = np.nonzero(target_loc)[0]
    zero_pos_list = np.random.choice(pos_list, size=int(len(pos_list) * 0.4), replace=False)
    target_loc[zero_pos_list] = 0
    target_loc = target_loc.astype(bool)
    data_df.loc[target_loc, 'is_click'] = 1
    return data_df

def process_data(data_df: pd.DataFrame):
    data_df['product_category_2'] = data_df[['product','campaign_id','webpage_id','product_category_2']].apply(handleProductCatagory2, axis = 1)
    data_df['gender'] = data_df[['product_category_1','gender']].apply(handleGender,axis = 1)
    data_df['city'] = data_df['city'].map({'Beijing': 0, 'Shanghai': 1, 'Guangzhou': 2, 'Shenzhen': 3, 'Hangzhou': 4})
    data_df['user_group_id'] = data_df[['gender','user_group_id']].apply(handleUserGroupId,axis = 1)
    data_df['age_level'].fillna(3,inplace = True)
    data_df['user_depth'].fillna(3,inplace = True)
    data_df['TimePart'] = data_df['DateTime'].apply(lambda x : x.split(' ')[1])
    data_df['DateDart'] = data_df['DateTime'].apply(lambda x : x.split(' ')[0])
    data_df.drop(['DateTime'],inplace=True,axis=1)
    data_df['Year'] = data_df['DateDart'].apply( lambda x : x.split('/')[2] )
    data_df['Month'] = data_df['DateDart'].apply( lambda x : x.split('/')[0] )
    data_df['Date'] = data_df['DateDart'].apply( lambda x : x.split('/')[1] )
    data_df['Hour'] = data_df['TimePart'].apply( lambda x : x.split(':')[0] )
    data_df['Minutes'] = data_df['TimePart'].apply( lambda x : x.split(':')[1] )
    data_df.drop(['DateDart'],axis = 1,inplace=True)
    data_df.drop(['TimePart'],axis = 1,inplace=True)
    gender = pd.get_dummies(data_df['gender'], drop_first=True)
    product_type = pd.get_dummies(data_df['product'])
    data_df.drop(['gender'],axis = 1,inplace=True)
    data_df.drop(['product'],axis = 1,inplace = True)
    data_df = pd.concat([data_df, product_type, gender],axis=1)
    return data_df

data: pd.DataFrame = pd.read_csv('click_log.csv')
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

data_train_raw = data[:350000].reset_index(drop=True)
data_test_raw = data[350000:].reset_index(drop=True)
print('mock data...')
data_test_raw = mock_new_year_eve_data(data_test_raw.copy())

print('process data...')
data_train = process_data(data_train_raw.copy())
data_test = process_data(data_test_raw.copy())

X_train = data_train.drop(['is_click'], axis=1)
y_train = data_train['is_click']
X_test = data_test.drop(['is_click'], axis=1)
y_test = data_test['is_click']
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
pred_test = dtc.predict(X_test)
print(confusion_matrix(y_test, pred_test))
print(classification_report(y_test, pred_test))
print(accuracy_score(y_test, pred_test))
```
